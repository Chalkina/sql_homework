```python
def remove_trash(list):
    pattern = r'[^a-zA-Z0-9]+'
    try:
        list = [re.sub(pattern, ' ', i) for i in list]
        print (list)
    except Exception as e:
        print(e)
    return list
```

```python
df['Name_clean'] = remove_trash(df['Name'])
```

```python
df = df[~df.Name_clean.isna()].copy()
```

```python
text_transformer = TfidfVectorizer(stop_words=stop_english, ngram_range=(1,1), lowercase=True, max_features=10000)
text = text_transformer.fit_transform(df['Name_clean'])
X_train, X_test, y_train, y_test = train_test_split(text, df['Sex'], test_size=0.25, random_state=1903)
```

```python
clf3 = RandomForestClassifier()
clf3.fit(X_train,y_train)
```

```python
pred3 = clf3.predict(X_test)
print('Качество модели по метрике F1 (токенизатор настроен и убран мусор)', f1_score(y_test,pred3,average='weighted'))
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
def lemmatize(row):
    t = []
    Name = row['Name_clean']
    for word in Name.split():
        if len(word)<=3:
            continue
        p = morph.parse(word)[0]
        t.append(p.normal_form)
    return " ".join(t)
```

```python
df['Name_clean_normal'] = df.apply(lemmatize,axis=1)
```

```python
Name_norm = text_transformer.fit_transform(df['Name_clean_normal'])
X_train, X_test, y_train, y_test = train_test_split(Name_norm, df['Sex'], test_size=0.25, random_state=1903)
```

```python
clf4 = RandomForestClassifier()
clf4.fit(X_train,y_train)
pred_norm = clf4.predict(X_test)
```

```python
print('Качество модели по метрике F1 после лемматизации', f1_score(y_test,pred_norm,average='weighted'))
```
