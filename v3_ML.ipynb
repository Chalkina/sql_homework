```python
data = pd.read_csv('/Users/Roman/Downloads/housing.csv')
pr = data['MEDV']
features = data.drop('MEDV', axis = 1)
print(features)
```

```python
df = pd.read_csv('/Users/Roman/Downloads/housing.csv')
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV',axis=1), df['MEDV'], test_size=0.2, random_state=42)
```

```python
X_train
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
print('Ошибка RMSE LinearRegression = ',sqrt(mean_squared_error(y_test,pred)))
```

```python
#Решающие деревья для задачи регрессии
```

```python
from sklearn import tree
tree_regressor = tree.DecisionTreeRegressor()
tree_regressor.fit(X_train,y_train)
```

```python
tree_pred = tree_regressor.predict(X_test)
```

```python
print('Ошибка RMSE DecisionTreeRegressor = ',sqrt(mean_squared_error(y_test,tree_pred)))
```

```python
frame = {'Важность параметров': tree_regressor.feature_importances_,'Параметры':X_train.columns}
result = pd.DataFrame(frame)
result.sort_values(by='Важность параметров', ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 9))
plt.plot(tree_pred, "r",linewidth=5, label="LinearRegression")
plt.plot(y_test.values,'b' ,label="actual", linewidth=5, alpha = 0.5)
plt.legend(loc="best")
plt.title("Оценка моделей")
plt.grid(True);
```
